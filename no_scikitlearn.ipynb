```python
## importing the pacages we nreversed
import pandas as pd
import numpy as numpy
import math
```

```python
x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 6, 8, 10]

def LinearRegression_1(X,Y,w=0.1,b=1,alpha=0.002,epochs=60,ws=[],bs=[],losses=[]): ## first with one feature
    ## first we go trough the epochs and compute an prediction
    for epoch in range(epochs):
        total_loss=0
        dw=0
        db=0
        m=len(X)
    
        for x,y in zip(X,X):
            y_hat = w*x+b ## the predict 1

            error = y_hat - y ## computing the erorr s2
            total_loss += error**2
        
            dw +=2 * error * x ## Compute gradients s3
            db +=2 * error

        w -= alpha * dw / m ## update the parameters s4
        b -= alpha * db / m
        
        avg_loss = total_loss / m
        ws.append(w)
        bs.append(b)
        losses.append(avg_loss)



    
    
    
    return ws,bs,losses

    
```

```python
w,b,l = LinearRegression_1(x_data,y_data)
```
